# Surg-RL: Proof of Concept Demonstration

This notebook demonstrates the core capabilities of Surg-RL for surgical robotics reinforcement learning.

## 1. Setup and Imports

````python
import sys
from pathlib import Path
import json
import time

project_root = Path('..').resolve()
sys.path.insert(0, str(project_root / 'src'))

from surg_rl.scene_definition import load_scene, validate_scene

print("✓ Imports successful")
````

## 2. Scene Loading and Validation

````python
scenes_dir = project_root / 'scenes'
scene_files = list(scenes_dir.glob('*.json')) + list(scenes_dir.glob('*.yaml'))

print(f"Found {len(scene_files)} scene files")
for scene_file in scene_files:
    print(f"  - {scene_file.name}")
````

````python
scene_file = scenes_dir / 'simple_suturing.json'
scene = load_scene(str(scene_file))
is_valid = validate_scene(scene)

print(f"✓ Scene: {scene.metadata.name}")
print(f"✓ Valid: {is_valid}")
````

## 3. Scene Analysis

````python
print(f"Robots: {len(scene.robots)}")
print(f"Tissues: {len(scene.tissues)}")
print(f"Instruments: {len(scene.instruments)}")
print(f"Task: {scene.task.name}")
````
